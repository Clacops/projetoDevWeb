```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configuração visual
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")

# Criando gráfico de barras para Custo Total por Linha de Cuidado
grafico = sns.barplot(
    data=analise_agrupada[analise_agrupada['Linha_Cuidado'] != 'Outros Diagnósticos'],
    x='VAL_TOT',
    y='Linha_Cuidado',
    palette='viridis'
)

plt.title('Custo Total Hospitalar por Linha de Cuidado (DATASUS - SC)')
plt.xlabel('Valor Total (R$)')
plt.ylabel('Linha de Cuidado')

# Adicionando rótulos de valores nas barras para facilitar a leitura
for i in grafico.containers:
    grafico.bar_label(i, fmt='R$%.2f', padding=5)

plt.show()
```

```python
# --- ANÁLISE ESTRATÉGICA ---
# Agrupamos por Linha de Cuidado para calcular média de permanência e custo total
analise_agrupada = df_estudo.groupby('Linha_Cuidado').agg({
    'DIAS_PERM': 'mean',
    'VAL_TOT': 'sum',
    'DIAG_PRINC': 'count'
}).rename(columns={'DIAG_PRINC': 'Total_Internacoes'}).reset_index()

# Ordenar pelas linhas com maior custo total (importante para saúde corporativa)
analise_agrupada = analise_agrupada.sort_values(by='VAL_TOT', ascending=False)

print("Relatório de Linhas de Cuidado (SC - Jan/2025):")
analise_agrupada.head()
```

```python
# Importamos as bibliotecas necessárias
from pysus.online_data.SIH import download
import pandas as pd
import numpy as np

# --- A. EXTRAÇÃO (AJUSTADA PARA PARQUETSET) ---
print("Iniciando download dos dados de SC... Aguarde.")
try:
    # Baixamos o conjunto de dados
    p_set = download('SC', 2025, 1, groups='RD')

    # CONVERSÃO: Transformamos o ParquetSet em um DataFrame do Pandas
    # Isso resolve o erro "object is not subscriptable"
    df_bruto = p_set.to_dataframe()

    print(f"Download e conversão concluídos! Total de registros: {len(df_bruto)}")
except Exception as e:
    print(f"Ocorreu um erro: {e}")

# --- B. SELEÇÃO DE COLUNAS ---
# Selecionamos colunas estratégicas para sua expertise em saúde e gestão [cite: 55, 57]
colunas_foco = ['DIAG_PRINC', 'IDADE', 'DIAS_PERM', 'VAL_TOT', 'MORTE']
df_estudo = df_bruto[colunas_foco].copy()

# --- C. TRATAMENTO DE TIPOS (Requisito da Vaga) ---
# Corrigindo falhas e inconsistências nos dados brutos
df_estudo['DIAS_PERM'] = pd.to_numeric(df_estudo['DIAS_PERM'], errors='coerce')
df_estudo['VAL_TOT'] = pd.to_numeric(df_estudo['VAL_TOT'], errors='coerce')

# --- D. CLASSIFICAÇÃO (Sua expertise clínica aplicada a dados) ---
def classificar_linha(cid):
    cid_str = str(cid).upper()
    if cid_str.startswith('I10'): return 'Hipertensão'
    if cid_str.startswith('E11'): return 'Diabetes'
    if cid_str.startswith('O80'): return 'Materno-Infantil'
    if cid_str.startswith('J44'): return 'DPOC/Respiratório'
    return 'Outros Diagnósticos'

df_estudo['Linha_Cuidado'] = df_estudo['DIAG_PRINC'].apply(classificar_linha)

# Visualização do resultado final
df_estudo.head(10)
```
